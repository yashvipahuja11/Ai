```python
import matplotlib.pyplot as plt
x=np.arange(0,10,1)
y= 2*x+5
y1=5*x+6
plt.plot(x,y,linewidth=2.0,linestyle=":", color='r',alpha=0.5,marker='o')
plt.plot(x,y1,linewidth=2.0,linestyle=":",color='b',alpha=0.9,marker='x')
plt.title("line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1','myline'],loc='best')
plt.grid(True)

plt.show() 
```

```python
x=np.arange(0,10,1)
y1=2*x+5
y2=3*x+10
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('graph2')
plt.show()
```
